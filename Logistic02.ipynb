```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('/content/logistic.xlsx')
df = df.values

x1 = df[:,0]
x2 = df[:,1]
x3 = df[:,2]
x4 = df[:,3]
y = df[:,4]
print("y :",y)

print("x1:",x1)
print("x2:",x2)
print("x3:",x3)
print("x4:",x4)

x1=x1.reshape((-1,1))
x2=x2.reshape((-1,1))
x3=x3.reshape((-1,1))
x4=x4.reshape((-1,1))

plt.scatter(x1,y)
plt.scatter(x2,y)
plt.scatter(x3,y)
plt.scatter(x4,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR1=LinearRegression()
SLR1.fit(x1,y)
SLR2=LinearRegression()
SLR2.fit(x2,y)
SLR3=LinearRegression()
SLR3.fit(x3,y)
SLR4=LinearRegression()
SLR4.fit(x4,y)
pred1=SLR1.predict(x1)
pred2=SLR2.predict(x1)
pred3=SLR3.predict(x3)
pred4=SLR4.predict(x4)
from sklearn.metrics import mean_squared_error
mse1=mean_squared_error(y,pred1)
mse2=mean_squared_error(y,pred2)
mse3=mean_squared_error(y,pred3)
mse4=mean_squared_error(y,pred4)
print("mse: ",mse1,mse2,mse3,mse4)
print("INTERCEPT :" ,SLR1.intercept_,SLR2.intercept_,SLR3.intercept_,SLR4.intercept_)
print("slope :",SLR1.coef_,SLR2.coef_,SLR3.coef_,SLR4.coef_)
plt.plot(x1,pred1,color='red',marker='*')
plt.plot(x2,pred2,color='orange',marker='.')
plt.plot(x3,pred3,color='blue',marker='+')
plt.plot(x4,pred4,color='green',marker='*')
```
